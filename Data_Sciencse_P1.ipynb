```python
import os
import pandas as pd

# List the extracted files
extracted_files = os.listdir()
print(extracted_files)

# Load the dataset (assuming the CSV file is named 'Hackathon_Ideal_Data.csv')
df = pd.read_csv('/content/Hackathon_Ideal_Data.csv')

# Display basic information about the dataset
df.info()
df.head()
```

```python
# Convert MONTH to a more interpretable format
df['MONTH'] = df['MONTH'].str.extract('(\d+)').astype(int)

# Display the first few rows to verify changes
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics for QTY and VALUE
summary_stats = df[['QTY', 'VALUE']].describe()
print(summary_stats)

# Histograms for QTY and VALUE
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['QTY'], bins=50, kde=True)
plt.title('Distribution of Quantity Sold (QTY)')
plt.xlabel('Quantity')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['VALUE'], bins=50, kde=True)
plt.title('Distribution of Transaction Value (VALUE)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Box plots for QTY and VALUE
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['QTY'])
plt.title('Box Plot of Quantity Sold (QTY)')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['VALUE'])
plt.title('Box Plot of Transaction Value (VALUE)')

plt.tight_layout()
plt.show()

# Bar charts for categorical columns
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.countplot(y=df['STORECODE'], order=df['STORECODE'].value_counts().index)
plt.title('Count of Transactions per Store')
plt.xlabel('Count')
plt.ylabel('Store Code')

plt.subplot(2, 2, 2)
sns.countplot(y=df['GRP'], order=df['GRP'].value_counts().index)
plt.title('Distribution of Product Groups (GRP)')
plt.xlabel('Count')
plt.ylabel('Group')

plt.subplot(2, 2, 3)
sns.countplot(y=df['CMP'], order=df['CMP'].value_counts().index)
plt.title('Distribution of Companies (CMP)')
plt.xlabel('Count')
plt.ylabel('Company')

plt.subplot(2, 2, 4)
sns.countplot(y=df['MBRD'], order=df['MBRD'].value_counts().index)
plt.title('Distribution of Main Brands (MBRD)')
plt.xlabel('Count')
plt.ylabel('Main Brand')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(y=df['STORECODE'], order=df['STORECODE'].value_counts().index)
plt.title('Count of Transactions per Store')
plt.xlabel('Count')
plt.ylabel('Store Code')
plt.show()

store_performance = df.groupby('STORECODE')[['QTY', 'VALUE']].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(x='STORECODE', y='VALUE', data=store_performance)
plt.title('Total Sales Value per Store')
plt.xlabel('Store Code')
plt.ylabel('Total Sales Value')
plt.show()
```

```python
monthly_sales = df.groupby('MONTH')['VALUE'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='MONTH', y='VALUE', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Value')
plt.show()
```

```python
from sklearn.cluster import KMeans

features = df[['QTY', 'VALUE']]
kmeans = KMeans(n_clusters=3)
kmeans.fit(features)
df['Cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
sns.scatterplot(x='QTY', y='VALUE', hue='Cluster', data=df, palette='viridis')
plt.title('Customer Segments based on Quantity and Value')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.show()
```
